```python
import matplotlib.pyplot as plt
plt.rc('font', size=16)
plt.rcParams['figure.constrained_layout.use'] = False
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# read in data
df = pd.read_csv('week5.csv')
print(df.head())
X=df.iloc[:,0]
y=df.iloc[:,1]

plt.xlabel('first training feature')
plt.ylabel('second training feature')
plt.scatter(X,y,label="training data",color="red")
plt.legend()
plt.show()


m = len(X)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_test=np.linspace(-3,3,num=len(y_test)).reshape(-1,1)
```

```python
from sklearn.neighbors import KNeighborsRegressor

# iterate over gamma values and train different models
gammas = [0,1,5,10,25]
for gamma in gammas:
    def gaussian_kernel(distances):
        weights = np.exp(-gamma*(distances**2))
        return weights/np.sum(weights)

    model_i = KNeighborsRegressor(n_neighbors=len(X_train),weights=gaussian_kernel).fit(X_train.values.reshape(-1,1), y_train)
    y_pred_i = model_i.predict(X_test)
    label_pred = "Predicted kNN gamma = " + str(gamma)
    plt.plot(X_test, y_pred_i, label=label_pred)
    plt.xlabel("input x"); plt.ylabel("output y")

plt.scatter(X_train, y_train, color="red", marker="+", label="training data")
plt.title("kNN Model with Gaussian Kernel")
plt.legend()
plt.show()
```

```python
from sklearn.kernel_ridge import KernelRidge

fig = plt.figure()
ax = fig.add_subplot(221) 

# iterate over gamma values and train different models
gammas = [0,1,5,10,25]
for gamma in gammas:
    c_values = [0.1,1,1000]
    plt.scatter(X_train, y_train, color="red", marker="+", label="training data")
    for c in c_values:
        model_i = KernelRidge(alpha=1.0/c, kernel="rbf", gamma=gamma).fit(X_train.values.reshape(-1,1), y_train)
        y_pred_i = model_i.predict(X_test)
        label_pred = "Predicted Kernel Ridge c= " + str(c) + " gamma=" + str(gamma)
        plt.plot(X_test, y_pred_i,label=label_pred)
        plt.xlabel("input x"); plt.ylabel("output y")
        print("Parameter values when c= " + str(c) + " gamma=" + str(gamma))
        print(model_i.dual_coef_)
    
    # Get the bounding box of the original legend
    legend = plt.legend( loc = 'upper right')
    bb = legend.get_bbox_to_anchor().inverse_transformed(ax.transAxes)

    # Change to location of the legend. 
    xOffset = 2
    bb.x0 += xOffset
    bb.x1 += xOffset
    legend.set_bbox_to_anchor(bb, transform = ax.transAxes)
    plt.title("Kernel Ridge gamma="+str(gamma))
    plt.show()
```

# Use cross validation to pick gamma for kNN Regressor


```python

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold

gammas = [0,1,5,10,25]
std_devs = []
means = []
for gamma in gammas:

    def gaussian_kernel(distances):
            weights = np.exp(-gamma*(distances**2))
            return weights/np.sum(weights)
        
    kf = KFold(n_splits=5)
    c = 1
    mse_estimates = []
    for train, test in kf.split(X):
        model_i = KNeighborsRegressor(n_neighbors=len(X[train]),weights=gaussian_kernel).fit(X[train].values.reshape(-1,1), y[train])
        y_pred_i = model_i.predict(X[test].values.reshape(-1,1))
        mse_estimates.append(mean_squared_error(y_pred_i, y[test]))

    means.append(np.mean(mse_estimates))
    std_devs.append(np.std(mse_estimates))

print(means)
print(std_devs)
plt.errorbar(gammas,means,yerr=std_devs,linewidth=3,capsize=5)
plt.title('KNeighborsRegressor for different gammas')
plt.xlabel('Gamma'); 
plt.ylabel('Mean/STD dev value')
plt.show()
```

# Use cross validation to pick gamma for Kernalised ridge regression

```python

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold

gammas = [0.1,1,5,10,25,50,100]
std_devs = []
means = []
for gamma in gammas:
    kf = KFold(n_splits=5)
    c = 1
    mse_estimates = []
    for train, test in kf.split(X):
        model_i = KernelRidge(alpha=1.0/1, kernel="rbf", gamma=gamma).fit(X[train].values.reshape(-1,1), y[train])
        y_pred_i = model_i.predict(X[test].values.reshape(-1,1))
        mse_estimates.append(mean_squared_error(y_pred_i, y[test]))

    means.append(np.mean(mse_estimates))
    std_devs.append(np.std(mse_estimates))

print(means)
print(std_devs)
plt.errorbar(gammas,means,yerr=std_devs,linewidth=3,capsize=5)
plt.title('Kernal Ridge for different gammas')
plt.xlabel('gamma'); 
plt.ylabel('Mean/STD dev value')
plt.show()
```

# Use cross validaiton to pick alpha for kernal ridge
